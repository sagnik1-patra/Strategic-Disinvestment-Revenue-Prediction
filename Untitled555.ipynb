```python
import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

# =====================================================
# 1ï¸âƒ£ PATH
# =====================================================

base_path = r"C:\Users\NXTWAVE\Downloads\Strategic Disinvestment Revenue Prediction"
file_path = os.path.join(base_path, "fetchOGDReceiptStrategicSale.json")

# =====================================================
# 2ï¸âƒ£ LOAD DATA
# =====================================================

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# =====================================================
# 3ï¸âƒ£ CLEAN DATA
# =====================================================

df['amountRealised'] = df['amountRealised'].astype(str)
df['amountRealised'] = df['amountRealised'].str.replace(",", "")
df['amountRealised'] = df['amountRealised'].str.replace("*", "")
df['amountRealised'] = pd.to_numeric(df['amountRealised'], errors='coerce')

df['equitySold'] = pd.to_numeric(df['equitySold'], errors='coerce')
df['year_clean'] = df['year'].str.extract(r'(\d{4})')
df['year_clean'] = pd.to_numeric(df['year_clean'], errors='coerce')

df = df.dropna(subset=['amountRealised','equitySold','year_clean'])

print("\nðŸ“Š Cleaned Dataset Shape:", df.shape)

# =====================================================
# 4ï¸âƒ£ HEATMAP (SHOW + SAVE)
# =====================================================

plt.figure(figsize=(6,5))
sns.heatmap(df[['year_clean','equitySold','amountRealised']].corr(),
            annot=True, cmap='coolwarm')
plt.title("PIS Hybrid Heatmap")
plt.savefig(os.path.join(base_path,"pis_heatmap.png"))
plt.show()

# =====================================================
# 5ï¸âƒ£ FEATURES
# =====================================================

X = df[['year_clean','equitySold']]
y = df['amountRealised']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# =====================================================
# 6ï¸âƒ£ AIS â€“ Feature Weight Optimization
# =====================================================

def ais_optimize(X_train, y_train):
    best_score = -999
    best_weights = None
    
    for _ in range(20):
        weights = np.random.uniform(0.5,1.5,size=X_train.shape[1])
        X_weighted = X_train * weights
        
        model = RandomForestRegressor(n_estimators=100)
        model.fit(X_weighted,y_train)
        pred = model.predict(X_weighted)
        score = r2_score(y_train,pred)
        
        if score > best_score:
            best_score = score
            best_weights = weights
    
    return best_weights

feature_weights = ais_optimize(X_train,y_train)

X_train_ais = X_train * feature_weights
X_test_ais = X_test * feature_weights

# =====================================================
# 7ï¸âƒ£ PSO â€“ Hyperparameter Optimization
# =====================================================

def pso_optimize(X_train,y_train):
    particles = 20
    best_global_score = -999
    best_global_params = None
    
    for _ in range(particles):
        n_estimators = random.randint(100,300)
        max_depth = random.randint(3,20)
        
        model = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth
        )
        model.fit(X_train,y_train)
        pred = model.predict(X_train)
        score = r2_score(y_train,pred)
        
        if score > best_global_score:
            best_global_score = score
            best_global_params = (n_estimators,max_depth)
    
    return best_global_params

best_n, best_depth = pso_optimize(X_train_ais,y_train)

# =====================================================
# 8ï¸âƒ£ FINAL HYBRID MODEL
# =====================================================

pis_model = RandomForestRegressor(
    n_estimators=best_n,
    max_depth=best_depth
)

pis_model.fit(X_train_ais,y_train)

y_pred = pis_model.predict(X_test_ais)

pis_r2 = r2_score(y_test,y_pred)
pis_rmse = np.sqrt(mean_squared_error(y_test,y_pred))

print("\nðŸ”¥ PIS (AIS+PSO) RESULTS")
print("Best n_estimators:", best_n)
print("Best max_depth:", best_depth)
print("R2 Score:", pis_r2)
print("RMSE:", pis_rmse)

# =====================================================
# 9ï¸âƒ£ SAVE MODEL
# =====================================================

joblib.dump(pis_model,
            os.path.join(base_path,"pis_model.pkl"))

# =====================================================
# ðŸ”Ÿ ACCURACY GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["PIS Hybrid"],[pis_r2])
plt.title("PIS Model Accuracy (R2)")
plt.ylabel("R2 Score")
plt.savefig(os.path.join(base_path,"pis_accuracy_graph.png"))
plt.show()

# =====================================================
# 11ï¸âƒ£ RMSE GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["PIS Hybrid"],[pis_rmse])
plt.title("PIS Model RMSE")
plt.ylabel("RMSE")
plt.savefig(os.path.join(base_path,"pis_rmse_graph.png"))
plt.show()

# =====================================================
# 12ï¸âƒ£ SAVE RESULT CSV
# =====================================================

results = pd.DataFrame({
    "Model":["PIS (AIS+PSO)"],
    "R2":[pis_r2],
    "RMSE":[pis_rmse],
    "Best_n_estimators":[best_n],
    "Best_max_depth":[best_depth]
})

results_path = os.path.join(base_path,"pis_results.csv")
results.to_csv(results_path,index=False)

print("\nðŸ“ Results saved at:", results_path)
print(results)

# =====================================================
# 13ï¸âƒ£ SAVE PREDICTION CSV
# =====================================================

pred_df = pd.DataFrame({
    "Actual":y_test.values,
    "PIS_Predicted":y_pred
})

pred_path = os.path.join(base_path,"pis_prediction.csv")
pred_df.to_csv(pred_path,index=False)

print("\nðŸ“ Prediction saved at:", pred_path)
print(pred_df.head())

# =====================================================
# 14ï¸âƒ£ PREDICTION GRAPH
# =====================================================

plt.figure(figsize=(7,6))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Revenue")
plt.ylabel("PIS Predicted Revenue")
plt.title("PIS Hybrid Prediction")
plt.savefig(os.path.join(base_path,"pis_prediction_graph.png"))
plt.show()

print("\nâœ… PIS HYBRID MODEL COMPLETED & SAVED SUCCESSFULLY!")
```
