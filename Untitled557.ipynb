```python
import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

# =====================================================
# 1ï¸âƒ£ PATH
# =====================================================

base_path = r"C:\Users\NXTWAVE\Downloads\Strategic Disinvestment Revenue Prediction"
file_path = os.path.join(base_path, "fetchOGDReceiptStrategicSale.json")

# =====================================================
# 2ï¸âƒ£ LOAD DATA
# =====================================================

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# =====================================================
# 3ï¸âƒ£ CLEAN DATA
# =====================================================

df['amountRealised'] = df['amountRealised'].astype(str)
df['amountRealised'] = df['amountRealised'].str.replace(",", "")
df['amountRealised'] = df['amountRealised'].str.replace("*", "")
df['amountRealised'] = pd.to_numeric(df['amountRealised'], errors='coerce')

df['equitySold'] = pd.to_numeric(df['equitySold'], errors='coerce')
df['year_clean'] = df['year'].str.extract(r'(\d{4})')
df['year_clean'] = pd.to_numeric(df['year_clean'], errors='coerce')

df = df.dropna(subset=['amountRealised','equitySold','year_clean'])

print("\nðŸ“Š Cleaned Dataset Shape:", df.shape)

# =====================================================
# 4ï¸âƒ£ HEATMAP
# =====================================================

plt.figure(figsize=(6,5))
sns.heatmap(df[['year_clean','equitySold','amountRealised']].corr(),
            annot=True, cmap='coolwarm')
plt.title("GIS Hybrid Heatmap")
plt.savefig(os.path.join(base_path,"gis_heatmap.png"))
plt.show()

# =====================================================
# 5ï¸âƒ£ FEATURES
# =====================================================

X = df[['year_clean','equitySold']]
y = df['amountRealised']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# =====================================================
# 6ï¸âƒ£ AIS â€“ Feature Weight Optimization
# =====================================================

def ais_optimize(X_train, y_train):
    best_score = -999
    best_weights = None
    
    for _ in range(20):
        weights = np.random.uniform(0.5,1.5,size=X_train.shape[1])
        X_weighted = X_train * weights
        
        model = RandomForestRegressor(n_estimators=100)
        model.fit(X_weighted,y_train)
        pred = model.predict(X_weighted)
        score = r2_score(y_train,pred)
        
        if score > best_score:
            best_score = score
            best_weights = weights
    
    return best_weights

feature_weights = ais_optimize(X_train,y_train)

X_train_ais = X_train * feature_weights
X_test_ais = X_test * feature_weights

# =====================================================
# 7ï¸âƒ£ GA â€“ Genetic Algorithm (Hyperparameter Optimization)
# =====================================================

def genetic_algorithm(X_train,y_train, generations=10, population_size=20):
    best_score = -999
    best_params = None
    
    for _ in range(generations):
        population = []
        for _ in range(population_size):
            n_estimators = random.randint(100,400)
            max_depth = random.randint(3,25)
            population.append((n_estimators,max_depth))
        
        for params in population:
            model = RandomForestRegressor(
                n_estimators=params[0],
                max_depth=params[1]
            )
            model.fit(X_train,y_train)
            pred = model.predict(X_train)
            score = r2_score(y_train,pred)
            
            if score > best_score:
                best_score = score
                best_params = params
    
    return best_params

best_n, best_depth = genetic_algorithm(X_train_ais,y_train)

# =====================================================
# 8ï¸âƒ£ FINAL GIS MODEL
# =====================================================

gis_model = RandomForestRegressor(
    n_estimators=best_n,
    max_depth=best_depth
)

gis_model.fit(X_train_ais,y_train)

y_pred = gis_model.predict(X_test_ais)

gis_r2 = r2_score(y_test,y_pred)
gis_rmse = np.sqrt(mean_squared_error(y_test,y_pred))

print("\nðŸ”¥ GIS (GA+AIS) RESULTS")
print("Best n_estimators:", best_n)
print("Best max_depth:", best_depth)
print("R2 Score:", gis_r2)
print("RMSE:", gis_rmse)

# =====================================================
# 9ï¸âƒ£ SAVE MODEL
# =====================================================

joblib.dump(gis_model,
            os.path.join(base_path,"gis_model.pkl"))

# =====================================================
# ðŸ”Ÿ ACCURACY GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["GIS Hybrid"],[gis_r2])
plt.title("GIS Model Accuracy (R2)")
plt.ylabel("R2 Score")
plt.savefig(os.path.join(base_path,"gis_accuracy_graph.png"))
plt.show()

# =====================================================
# 11ï¸âƒ£ RMSE GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["GIS Hybrid"],[gis_rmse])
plt.title("GIS Model RMSE")
plt.ylabel("RMSE")
plt.savefig(os.path.join(base_path,"gis_rmse_graph.png"))
plt.show()

# =====================================================
# 12ï¸âƒ£ SAVE RESULT CSV
# =====================================================

results = pd.DataFrame({
    "Model":["GIS (GA+AIS)"],
    "R2":[gis_r2],
    "RMSE":[gis_rmse],
    "Best_n_estimators":[best_n],
    "Best_max_depth":[best_depth]
})

results_path = os.path.join(base_path,"gis_results.csv")
results.to_csv(results_path,index=False)

print("\nðŸ“ Results saved at:", results_path)
print(results)

# =====================================================
# 13ï¸âƒ£ SAVE PREDICTION CSV
# =====================================================

pred_df = pd.DataFrame({
    "Actual":y_test.values,
    "GIS_Predicted":y_pred
})

pred_path = os.path.join(base_path,"gis_prediction.csv")
pred_df.to_csv(pred_path,index=False)

print("\nðŸ“ Prediction saved at:", pred_path)
print(pred_df.head())

# =====================================================
# 14ï¸âƒ£ PREDICTION GRAPH
# =====================================================

plt.figure(figsize=(7,6))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Revenue")
plt.ylabel("GIS Predicted Revenue")
plt.title("GIS Hybrid Prediction")
plt.savefig(os.path.join(base_path,"gis_prediction_graph.png"))
plt.show()

print("\nâœ… GIS HYBRID MODEL COMPLETED & SAVED SUCCESSFULLY!")
```
