```python
import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

# =====================================================
# 1️⃣ PATH
# =====================================================

base_path = r"C:\Users\NXTWAVE\Downloads\Strategic Disinvestment Revenue Prediction"
file_path = os.path.join(base_path, "fetchOGDReceiptStrategicSale.json")

# =====================================================
# 2️⃣ LOAD DATA
# =====================================================

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# =====================================================
# 3️⃣ CLEAN DATA
# =====================================================

df['amountRealised'] = df['amountRealised'].astype(str)
df['amountRealised'] = df['amountRealised'].str.replace(",", "")
df['amountRealised'] = df['amountRealised'].str.replace("*", "")
df['amountRealised'] = pd.to_numeric(df['amountRealised'], errors='coerce')

df['equitySold'] = pd.to_numeric(df['equitySold'], errors='coerce')
df['year_clean'] = df['year'].str.extract(r'(\d{4})')
df['year_clean'] = pd.to_numeric(df['year_clean'], errors='coerce')

df = df.dropna(subset=['amountRealised','equitySold','year_clean'])

print("\n📊 Cleaned Dataset Shape:", df.shape)

# =====================================================
# 4️⃣ HEATMAP
# =====================================================

plt.figure(figsize=(6,5))
sns.heatmap(df[['year_clean','equitySold','amountRealised']].corr(),
            annot=True, cmap='coolwarm')
plt.title("SWLOA Heatmap")
plt.savefig(os.path.join(base_path,"swloa_heatmap.png"))
plt.show()

# =====================================================
# 5️⃣ FEATURES
# =====================================================

X = df[['year_clean','equitySold']]
y = df['amountRealised']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# =====================================================
# 6️⃣ ALOA — Global Search
# =====================================================

def aloa_search(X_train,y_train,agents=25):
    best_score = -999
    best_params = None
    
    for _ in range(agents):
        n_estimators = random.randint(100,500)
        max_depth = random.randint(3,35)
        
        model = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth
        )
        model.fit(X_train,y_train)
        pred = model.predict(X_train)
        score = r2_score(y_train,pred)
        
        if score > best_score:
            best_score = score
            best_params = (n_estimators,max_depth)
    
    return best_params

best_n, best_depth = aloa_search(X_train,y_train)

print("\n🔹 ALOA Best Params:", best_n, best_depth)

# =====================================================
# 7️⃣ WOA — Refinement Phase
# =====================================================

def woa_refine(X_train,y_train,base_n,base_depth,iterations=20):
    best_score = -999
    best_params = (base_n, base_depth)
    
    for t in range(iterations):
        a = 2 - t*(2/iterations)
        
        A = 2*a*random.random() - a
        C = 2*random.random()
        
        D_n = abs(C*base_n - base_n)
        D_d = abs(C*base_depth - base_depth)
        
        new_n = int(base_n - A*D_n)
        new_d = int(base_depth - A*D_d)
        
        new_n = max(50, new_n)
        new_d = max(2, new_d)
        
        model = RandomForestRegressor(
            n_estimators=new_n,
            max_depth=new_d
        )
        model.fit(X_train,y_train)
        pred = model.predict(X_train)
        score = r2_score(y_train,pred)
        
        if score > best_score:
            best_score = score
            best_params = (new_n,new_d)
    
    return best_params

final_n, final_depth = woa_refine(X_train,y_train,best_n,best_depth)

print("🔹 WOA Refined Params:", final_n, final_depth)

# =====================================================
# 8️⃣ FINAL MODEL
# =====================================================

swloa_model = RandomForestRegressor(
    n_estimators=final_n,
    max_depth=final_depth
)

swloa_model.fit(X_train,y_train)

y_pred = swloa_model.predict(X_test)

swloa_r2 = r2_score(y_test,y_pred)
swloa_rmse = np.sqrt(mean_squared_error(y_test,y_pred))

print("\n🔥 SWLOA RESULTS")
print("Final n_estimators:", final_n)
print("Final max_depth:", final_depth)
print("R2 Score:", swloa_r2)
print("RMSE:", swloa_rmse)

# =====================================================
# 9️⃣ SAVE MODEL
# =====================================================

joblib.dump(swloa_model,
            os.path.join(base_path,"swloa_model.pkl"))

# =====================================================
# 🔟 ACCURACY GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["SWLOA Hybrid"],[swloa_r2])
plt.title("SWLOA Model Accuracy (R2)")
plt.ylabel("R2 Score")
plt.savefig(os.path.join(base_path,"swloa_accuracy_graph.png"))
plt.show()

# =====================================================
# 11️⃣ RMSE GRAPH
# =====================================================

plt.figure(figsize=(6,5))
plt.bar(["SWLOA Hybrid"],[swloa_rmse])
plt.title("SWLOA Model RMSE")
plt.ylabel("RMSE")
plt.savefig(os.path.join(base_path,"swloa_rmse_graph.png"))
plt.show()

# =====================================================
# 12️⃣ SAVE RESULT CSV
# =====================================================

results = pd.DataFrame({
    "Model":["SWLOA (ALOA+WOA)"],
    "R2":[swloa_r2],
    "RMSE":[swloa_rmse],
    "Final_n_estimators":[final_n],
    "Final_max_depth":[final_depth]
})

results_path = os.path.join(base_path,"swloa_results.csv")
results.to_csv(results_path,index=False)

print("\n📁 Results saved at:", results_path)
print(results)

# =====================================================
# 13️⃣ SAVE PREDICTION CSV
# =====================================================

pred_df = pd.DataFrame({
    "Actual":y_test.values,
    "SWLOA_Predicted":y_pred
})

pred_path = os.path.join(base_path,"swloa_prediction.csv")
pred_df.to_csv(pred_path,index=False)

print("\n📁 Prediction saved at:", pred_path)
print(pred_df.head())

# =====================================================
# 14️⃣ PREDICTION GRAPH
# =====================================================

plt.figure(figsize=(7,6))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Revenue")
plt.ylabel("SWLOA Predicted Revenue")
plt.title("SWLOA Hybrid Prediction")
plt.savefig(os.path.join(base_path,"swloa_prediction_graph.png"))
plt.show()

print("\n✅ SWLOA HYBRID MODEL COMPLETED & SAVED SUCCESSFULLY!")
```
